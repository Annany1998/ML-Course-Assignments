```python
df1 = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'], 'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
# Q2.
```

```python
def From_col(x):
    From=[]
    temp=x.split('_')
    From.append(temp[0])
    return From[0]
def To_col(x):
    To=[]
    temp=x.split('_')
    To.append(temp[1])
    return To[0]
```

```python
df1['From']=df1['From_To'].apply(From_col)
df1['To']=df1['From_To'].apply(To_col)
```

```python
df1
```

```python
# Q3.
```

```python
def str_prob(x):
    x=x.lower()
    temp=[]
    if x[0].islower():
        temp.append(x[0].upper())
    for i in range(1,len(x)):
        temp.append(x[i])
    s=''
    return (s.join(temp))     
```

```python
df1['From']=df1['From'].apply(str_prob)
```

```python
df1['To']=df1['To'].apply(str_prob)
```

```python
delays=pd.DataFrame(df1,copy=True)
```

```python
def delay1(lst):
    if lst==[]:
        return None
    else:
        return lst[0]
def delay2(lst):
    if len(lst)<2:
        return None
    else:        
        return lst[1]
def delay3(lst):
    if len(lst)<3:
        return None
    else:
        return lst[2]
```

```python
delays['delay_1']=delays['RecentDelays'].apply(delay1)
delays['delay_2']=delays['RecentDelays'].apply(delay2)
delays['delay_3']=delays['RecentDelays'].apply(delay3)
```

```python
delays.drop('RecentDelays',inplace=True,axis=1)
```
